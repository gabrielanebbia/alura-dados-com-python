```python
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

### Aplica renomeação
df.rename(columns=novos_nomes, inplace=True)
df.head(10)
```

```python
## Analisando categorias das colunas categóricas
```

```python
### Nível de senioridade
### O método para contar quantas vezes cada valor único aparece em uma coluna
df['senioridade'].value_counts()

#### Legenda dos níveis de senioridade
#### SE	Senior	Profissional experiente (nível sênior)
#### MI	Mid	Nível intermediário
#### EN	Entry	Iniciante ou júnior (entry-level)
#### EX	Executive	Executivo, liderança sênior (C-Level)
```

```python
### Tipo de contrato
df['contrato'].value_counts()

#### Legenda dos tipos de contrato
#### FT Full-time	Tempo integral – trabalho padrão de 40h/semana ou período completo
#### PT Part-time	Meio período – carga horária reduzida
#### CT Contract	Contrato temporário ou por projeto – geralmente por prazo determinado
#### FL Freelance	Freelancer – trabalho autônomo ou por demanda, sem vínculo formal
```

```python
### Regime de trabalho
df['remoto'].value_counts()

#### Legenda dos regimes de trabalho
#### 0 Presencial
#### 100 Remoto
#### 50 Híbrido
```

```python
### Cargo mais frequente
df['cargo'].value_counts()
```

```python
# País dos profissionais
df['residencia'].value_counts()
```

```python
# País de sede das empresas
df['empresa'].value_counts()
```

```python
# Salarios por moeda
df['moeda'].value_counts()
```

```python
# Salarios (média, min, max)
df['usd'].describe()
```

```python
## Modificando o nome das categorias
```

```python
senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M': 'media'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
mapa_trabalho = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}

df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['remoto'].value_counts()
```

```python
### Resume informações categóricas com o método describe(), exibindo a quantidade de categorias únicas, categoria mais frequente e respectiva frequência:
df.describe(include='object')
```

```python
### Para pandas versões 1.0.0 e superiores, que introduziram o dtype 'string', é possível incluir ambos os tipos textuais de duas formas:
#### 1. Inclui explicitamente ambos os dtypes textuais
df.describe(include=['object', 'string'])
```

```python
#### 2. Ou selecionar colunas textuais e descrever apenas elas
df.select_dtypes(include=['object', 'string']).describe()
```
